# Кластерный анализ
## Метод к-средних



Дана матрица данных $X$ и дано число $k$ предполагаемых кластеров. Цель кластеризации представить данные в виде групп кластеров $C=\{C_1, C_2, \ldots, C_k\}$. Каждый кластер имеет свой центр:

\begin{equation}
\mu_i = \frac{1}{n_i} \sum \limits_{x_j \in C_i} x_j
\end{equation}

где $n_i = |C_i|$ - это количество точек в кластере $C_i$.

Таким образом, мы имеем некоторые кластеры $C=\{C_1, C_2, \ldots, C_k\}$ и необходимо оценить качество разбиения. Для этого будем вычислять сумму квадратических ошибок (SSE):

\begin{equation}
SSE(C) = \sum \limits_{i=1}^{k} \sum \limits_{x_j \in C_i} ||x_j - \mu_i||^2
\end{equation}

Цель найти
\begin{equation}
C^* = arg\min\limits_C \{SSE(C)\}
\end{equation}

### Алгоритм к-средних
На вход алгоритм получает матрицу данных $D$, количество кластеров $k$, и критерий остановки $\epsilon$:
1. t = 0
2. случайным образом инициализируем $k$ центров кластеров: $\mu_1^t, \mu_2^t, \ldots, \mu_k^t \in R^d$;
3. повторять
    1. $t = t + 1$;
    2. $C_j = 0$ для всех $j = 1, \ldots, k$
    3. для каждого $x_j \in D$
        1. $j^* = arg\min\limits_C \{||X_j - \mu_i^{t-1}||^2\}$ \\\ присваиваем $x_j$ к ближайшему центру
        2. $C_{j^*} = C_{j^*} \cup {x_j}$
    4. для каждого i=1 до k
        1. $\mu_i = \frac{1}{|C_i|} \sum_{x_j \in C_i} x_j$
4. пока $\sum_{i=1}^k ||\mu_i^{t} - \mu_i^{t-1}||^2 \leq \epsilon$

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
X, Y = make_blobs(n_samples = 1000, n_features=2, centers=5, cluster_std = 1.2, random_state=17)
plt.scatter(X[:,0], X[:,1], c=Y)
```

```python
def d(q, p):
    return ((q - p) ** 2).sum() ** (1 / 2)
```

```python
e = .1e+0
```

```python
tries = 10
K = 6
```

```python
SSEk = np.zeros(K)
yhatk = []
mutk = []
for k in range(1, K+1):
    SSE = np.zeros(tries)
    yhat = []
    mutr = []
    for r in range(tries):
        # 2
#         mut = np.array((np.random.uniform(X[:,0].min(), X[:,0].max(), k),
#                       np.random.uniform(X[:,1].min(), X[:,1].max(), k)), copy=False).transpose()
        mui = np.random.choice(np.arange(X.shape[0]), k)
        mut = np.array([X[i] for i in mui])

        # 3
        statement = True
        while(statement):
            ds = np.array([[d(X[i], mut[j])**2 for j in range(k)] for i in range(X.shape[0])])
            yhat.append(np.array([np.squeeze(np.where(ds[i] == ds[i].min())) for i in range(ds.shape[0])]))
            ymap = [np.where(yhat[r] == i)[0] for i in range(k)]
            count = np.array([len(ymap[i]) for i in range(k)])
            count.shape = (k, 1)
            mutp = mut.copy()
            mut_sum = [np.sum([X[i] for i in ymap[j]], 0) for j in range(k)]
            for i, el in enumerate(mut_sum):
                if type(el) is np.float64:
                    mut_sum[i] = np.zeros(2)
            mut = np.array(mut_sum) / count

            # 4
            statement = np.sum([d(mut[i], mutp[i]) for i in range(mut.shape[0])]) <= e
        SSE[r] = np.sum([d(mut[i], X[j])**2 for i in range(k) for j in ymap[i]])
        mutr.append(mut)
    
    best_index = np.where(SSE == SSE.min())[0][0]
    mutk.append(mutr[best_index])
    yhatk.append(yhat[best_index])
    SSEk[k-1] = SSE.min()
```

```python
for i in range(len(yhatk)):
    plt.figure()
    plt.scatter(X[:,0], X[:,1], c=yhatk[i])
    plt.scatter(mutk[i][:,0], mutk[i][:,1])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv', encoding='cp1251')
df = df.drop(columns=['Модель', 'Цвет'])
df.head()
```

```python
new_df = pd.get_dummies(df)
new_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
new_df[['Год', 'Пробег', 'Объем', 'Мощность', 'Цена']] = ss.fit_transform(new_df[['Год', 'Пробег', 'Объем', 'Мощность', 'Цена']])
new_df.head()
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, AffinityPropagation
inertia = []
for it in np.arange(10,11,1):
    method = KMeans(n_clusters=it)
    method.fit(new_df)
    inertia.append(method.inertia_)
    print(it)
```

```python
def cls_info(df, k):
    print('------ Кластер ', k, ' -------')
    claster =  df[method.labels_ == k]
    print(claster['Год'].mean())
    print(claster['Пробег'].mean())
    print(claster['Объем'].mean())
    print(claster['Мощность'].mean())
    print(claster['Цена'].mean())
    print(claster['Привод'].value_counts().head(1))
    print(claster['Марка'].value_counts().head(2))
    print(claster['Кузов'].value_counts().head(2))
    print(claster['КПП'].value_counts().head(2))
    print('---------------------------')
```

```python
for it in range(10):
    cls_info(df, it)
```
